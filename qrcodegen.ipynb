```python
import qrcode
data = "https://www.youtube.com/watch?v=px0RUs1Iv2M"


qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_M,
    box_size=25,
    border = 1
)
#add the code to QRCode instance

qr.add_data(data)
#generate qrcode matrix
qr.make(fit=True)

# create an image from qrcode
image = qr.make_image(fill_color="black", back_color="white")

image.save("mychannelqrcode.png")
```
